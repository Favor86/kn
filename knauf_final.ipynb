# Preprocessing functions

```python
def f_knf(s):
    s=s.lower()
    if s.find('кнауф')+s.find('knauf')>-2: s='knauf '+ s.replace('knauf','').replace('кнауф','').strip().replace('  ',' ')
    return s

def f_repl(s):
    s=f_knf(s)
    s=s.replace('гклво','гипсокартонный лист влагоогнеустойчивый').replace('гкло','гипсокартонный лист огнеустойчивый')
    s=s.replace('гклв','гипсокартонный лист влагоустойчивый').replace('гкл','гипсокартонный лист')
    s=s.replace('гвлв','гипсоволокнистый лист влагоустойчивый').replace('гвл','гипсоволокнистый лист')
    s=s.replace('плук','полукруглая утоненная кромка')
    s=s.replace('д/','для ')
    s=s.replace('гсп-dfh3ir','гипсокартонный лист сапфир').replace('гсп-dfh2','гипсостроительная плита влагоогнеустойчивая')
    s=s.replace('гсп-h2','гипсостроительная плита влагоустойчивая').replace('гсп-df','гипсостроительная плита огнеустойчивая')
    s=s.replace('гсп','гипсостроительная плита').replace('ксл','комплектная система листов')
    if s.find('профиль')>-1:
        s=s.replace(' пу ',' профиль углозащитный ').replace(' пм ',' профиль маячковый ')
        s=s.replace(' па ',' профиль арочный ').replace(' ua ',' профиль усиленный ')
        s=s.replace(' ппн ',' профиль направляющий потолочный ').replace(' ппн ',' профиль направляющий потолочный ')
        s=s.replace(' пн ',' профиль направляющий ')
    #s=s.replace('','').replace('','')
    return s

def f_sizes(s):
    s=f_repl(s).replace('  ',' ')
    p='(\(| )(\d+((\.|,)\d*)? *(x|\*|х) *){1,3}\d+((\.|,)\d*)? ?' #размеры
    p2=' (\d+(\.|,)*\d+ *(\-) *){2,3}\d+(\.|,)*\d+ ?'
    p3=' (\d+((\.|,)\d*)? *)(x|\*|х) *(\d+((\.|,)\d*)? *) \d+((\.|,)\d*)? ?'
    p_kg='\d+ ?кг'
    if re.search(p3,s) is not None: #более частный случай с " " после 2х размеров
        return re.search(p3,s).group(0), 'size_x'
    elif re.search(p,s) is not None:
        return re.search(p,s).group(0), 'size_x'
    elif re.search(p2,s) is not None:
        return re.search(p2,s).group(0), 'size_x'

    elif re.search(p_kg,s) is not None:
        return re.search(p_kg,s).group(0), 'size_kg'
    else: return '',''

def re_razryad_replace(s): #иногда ставят запятые, чтобы отделить тысячи
    t_=re.search('(,|\.)\d\d0',s)
    if t_ is not None: return s.replace(t_.group(0)[0],'')
    return s
    
def f_sizes_ok(s,t):
    if t=='size_x':
        if s!='':
            sl=s.replace('(','').replace(',','.').replace('x',' ').replace('*',' ').replace('х',' ').replace('-',' ').split()
            return 'x'.join(list(map(str, sorted(list(map(float,list(map(re_razryad_replace,sl))
                                                         )),reverse=True))))#.replace('.0','')
    elif t=='size_kg':
        return s.replace(' ','')

def f_translit(s):
    #symbols = (u"абвгдеёжзийклмнопрстуфхцчшщъыьэюяАБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ",
    #           u"abvgdeejzijklmnoprstufhzcss_y_euaABVGDEEJZIJKLMNOPRSTUFHZCSS_Y_EUA")
    symbols = (u"abcdefghijklmnopqrstuvwxyz",
               u"абсдефгхийклмнопкрстуввхиз")
    tr = {ord(a):ord(b) for a, b in zip(*symbols)}
    return s.lower().translate(tr)


def f_final(s):
    s=f_repl(s)
    siz=f_sizes(s)[0]
    if siz!='':
        s=s.replace(siz,' <' + f_sizes_ok(siz,f_sizes(s)[1]) + '> ')
    
    knf=''
    dop=''
    if s.find('knauf ')>-1:
        s=s.replace('knauf ','')
        knf='knf'
    t_dop=re.search('\(\d+\)',s)
    if t_dop is not None:
        dop=t_dop.group(0)
        s=s.replace(dop,'').replace('  ',' ').replace('  ',' ').strip()
    s=f_translit(s)
    
    return knf, s, dop

```

# Pilot Rules 

```python
def pilot_find_by_rules(s, l, r):
        for i in l:
            and_2=i.split('+')
            cnt=0
            for j in and_2:
                or_3=j.split('|')
                or_3_yes=0
                for k in or_3:
                    if s.find(k)>-1: or_3_yes=1
                if or_3_yes==1: cnt=cnt+1
            #print(cnt,len(all_conditions))
            if cnt==len(and_2): return r
        return '-'
```

```python
def pilot_find_by_rules(s, l, r):
        for i in l:
            all_conditions=i.split('+')
            cnt=0
            for j in all_conditions:
                if s.find(j)>-1: cnt=cnt+1
            #print(cnt,len(all_conditions))
            if cnt==len(all_conditions): return r
        return '-'
    
def pilot_res_by_rules(s,rules_dict):
    for row in range(len(rules_dict)):
        l=rules_dict['list_r'].iloc[row]
        r=rules_dict['r'].iloc[row]
        res=pilot_find_by_rules(s, l, r)
        if pilot_find_by_rules(s, l, r)!='-': return res
    return ''
```

```python
def pilot_find(s):
    sl=s.lower()
    cat_i='-'
    cat_i=pilot_res_by_rules(sl,rules_dict)
    
    return cat_i
```
